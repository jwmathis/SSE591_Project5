```python
# Libraries
%matplotlib inline
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data as a table
iris = sns.load_dataset('iris')
iris.head()
# each row represents a single observed flower
# rows of the matrix are samples and the number of rows is n_samples
# columns are features that describe the each sample in a quantitative manner
# features matrix, X, is 2D with a shape of [n_samples, n_features] (a numpy array or pandas dataframe)
```

```python
# the target array, y, is 1D with length n_samples generally contianed in a numpy array or pandas series
# it is usually the quantity we want to predict form the data: i.e. the dependent variable
# for this example, we want to construct a modle that can predict the species of flower based on the other measuremnts

sns.pairplot(iris, hue='species', height=1.5);
```

```python
# features matrix without the target vector
X_iris = iris.drop('species', axis=1)
X_iris.shape
```

```python
# target vector
y_iris = iris['species']
y_iris.shape
```

### Supervised learning example: Simple linear regression

```python
# the data
rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x - 1 +rng.randn(50)
plt.scatter(x,y);
```

```python
# 1. Choose a class of model
from sklearn.linear_model import LinearRegression
# 2. Choose model hyperparameters
model = LinearRegression(fit_intercept=True)
model
```

```python
# 3. Arrange data into a features matrix and target vector
X = x[:, np.newaxis]
X.shape
```

```python
# Fit the model to your data
model.fit(X, y)
```

```python
# Slope
model.coef_
```

```python
# y-intercept
model.intercept_
```

```python
# 5. Predict labels for unknown data

## new data will be a grid of x-values for the sake of example
xfit = np.linspace(-1, 11)
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
# Visualize
plt.scatter(x, y)
plt.plot(xfit, yfit);
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris,
                                                random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB()                       # 2. instantiate model
model.fit(Xtrain, ytrain)                  # 3. fit model to data
y_model = model.predict(Xtest)             # 4. predict on new data
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```
